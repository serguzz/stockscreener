```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
# Define the words to exclude
all_words = {'Stock', 'stock', ' Shares', ' Share', ' shares', 'share', 'ProShares', 'Fund', ' ETF', 'ADS', 'Class A', 'Trust',
                  'Business Development Company', 'Preferred Class B Series 2 Reset Rate',
                  'Right', 'right', 'Unit', 'unit',
                  'Warrant', 'Warrants', 'warrant', 'Notes','NOTES', 'notes', 'TEST STOCK', 'Trust Preferred Securities'}

include_words = {'Stock', 'stock', ' Shares', ' Share', ' shares', 'share', 'ProShares', 'Fund', ' ETF', 'ADS', 'Class A', 'Trust',
                  'Business Development Company', 'Preferred Class B Series 2 Reset Rate'}

exclude_words = {'Right', 'right', 'Unit', 'unit', 'Warrant', 'Warrants', 'warrant', 'Notes','NOTES', 'notes',
                 'TEST STOCK', 'Trust Preferred Securities'}
```

```python
# Initialize an empty list to hold the symbols
symbols = []

# Read the contents of the file
with open('nasdaqlisted.txt', 'r') as file:
    # Fetch all lines from the file
    lines = file.readlines()
    print(len(lines))
    # Filter lines that contain "Stock"
    for line in lines:
        if any(word in line for word in include_words):  # and not any(word in line for word in exclude_words)
            # Get the symbol before the "|" character
            symbol = line.split('|')[0]
            symbols.append(symbol)

# Convert the list of symbols to a DataFrame
symbols_df = pd.DataFrame(symbols, columns=['Symbol'])

# Save the DataFrame to a CSV file
symbols_df.to_csv('nasdaq_symbols.csv', index=False)

# Print the list of symbols (optional)
print(len(symbols))
# print(symbols)
```

```python
# Initialize a list to hold stable symbols
symbols_stable = []
```

```python
# Define the date range to check for price data
# We are checking assets that are traded at least from 2017 
start_date = '2017-01-01'
end_date = '2017-01-20'
```

```python
# Iterate through the symbols
for symbol in symbols:
    # Fetch historical data for the last 8 years
    stock_data = yf.download(symbol, start=start_date, end=end_date)  # End date is exclusive
    
    # Check if the stock data is not empty and contains data for the specified dates
    if not stock_data.empty:
        symbols_stable.append(symbol)


# Save the stable symbols to a CSV file
stable_symbols_df = pd.DataFrame(symbols_stable, columns=['Symbol'])
stable_symbols_df.to_csv('nasdaq_symbols_stable.csv', index=False)

# Save the stable symbols to a variable
print("Stable Symbols:")
print(len(symbols_stable))
# print(symbols_stable)
```
