### This scipts changes float64 to float32 column types in OHLCV data in 'data/'folder

```python
import pandas as pd
import os
```

```python
ohlcv_folder = 'data/'
```

```python
def optimize_file(filename):
    """
    Loads a CSV file, converts columns with float64 data type to float32,
    and saves the optimized DataFrame back to the same file.
    """
    # Load the CSV file into a DataFrame
    df = pd.read_csv(filename)
    
    # Convert columns with float64 type to float32
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = df[col].astype('float32')
    
    # Save the optimized DataFrame back to the CSV file
    df.to_csv(filename, index=False)
    print(f"Optimized file saved: {filename}")
```

```python
# Get and sort all .csv files in alphabetical order
csv_files = sorted([file for file in os.listdir(ohlcv_folder) if file.endswith('.csv')])

# Iterate through all files in the folder
for file in csv_files:
    # Check if the file has a '.csv' extension
    filepath = os.path.join(ohlcv_folder, file)
    optimize_file(filepath)
```
