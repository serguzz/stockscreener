### The script uses nasdaqlisted endpoint to get and parse all Nasdaq traded stock and from them we choose only those stocks that have been traded not less than 8 years (i.e., have trading data in January 2016).
### The stocks tickers are saved in a csv file

```python
# We will use only  Nasdaq symbols from https://www.nasdaqtrader.com/dynamic/SymDir/nasdaqlisted.txt

# Define the URL for the Nasdaq symbol directory
# Page "https://www.nasdaqtrader.com/dynamic/SymDir/nasdaqlisted.txt"
# saved to "nasdaqlisted.txt"  - in this folder
```

```python
import pandas as pd

# Initialize an empty list to hold the symbols
symbols = []

# Read the contents of the file
with open('nasdaqlisted.txt', 'r') as file:
    # Fetch all lines from the file
    lines = file.readlines()
    
    # Filter lines that contain "Stock"
    for line in lines:
        if "Stock" in line:
            # Get the symbol before the "|" character
            symbol = line.split('|')[0]
            symbols.append(symbol)

# Save the symbols to a variable
# symbols now contains the list of stock symbols

# Convert the list of symbols to a DataFrame
symbols_df = pd.DataFrame(symbols, columns=['Symbol'])

# Save the DataFrame to a CSV file
symbols_df.to_csv('nasdaq_stocks.csv', index=False)

# Print the list of symbols (optional)
# print(symbols)
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Load the previously saved symbols from 'nasdaq_stocks.csv'
symbols_df = pd.read_csv('nasdaq_stocks.csv')
symbols = symbols_df['Symbol'].tolist()  # Convert to a list

# Initialize a list to hold stable symbols
symbols_stable = []

# Define the date range to check for price data
start_date = '2016-01-01'
end_date = '2016-01-20'

# Iterate through the symbols
for symbol in symbols:
    # Fetch historical data for the last 8 years
    stock_data = yf.download(symbol, start=start_date, end=end_date)  # End date is exclusive
    
    # Check if the stock data is not empty and contains data for the specified dates
    if not stock_data.empty:
        symbols_stable.append(symbol)


# Save the stable symbols to a CSV file
stable_symbols_df = pd.DataFrame(symbols_stable, columns=['Symbol'])
stable_symbols_df.to_csv('nasdaq_stocks_stable.csv', index=False)

# Save the stable symbols to a variable
print("Stable Symbols:")
print(len(symbols_stable))
print(symbols_stable)
```
